At the end of April 2020, a group of behavioural scientists [@bavel2020] published a paper in Nature Human Behaviour, "Using social and behavioural science to support COVID-19 pandemic response". It included a set of 19 recommendations for policy makers.

It was widely discussed and cited and was in the middle of some discussions about whether behavioural science was mature enough to guide the pandemic response (such as by ).

A new article in Nature [@ruggeri2023] has reviewed the claims in that 2020 paper. Part of the abstract reads:

> In April 2020, an influential paper proposed 19 policy recommendations (‘claims’) detailing how evidence from behavioural science could contribute to efforts to reduce impacts and end the COVID-19 pandemic. Here we assess 747 pandemic-related research articles that empirically investigated those claims. We report the scale of evidence and whether evidence supports them to indicate applicability for policymaking. Two independent teams, involving 72 reviewers, found evidence for 18 of 19 claims, with both teams finding evidence supporting 16 (89%) of those 18 claims.

Reading this narrowly, it's a tick for that 2020 paper.

But, I wondered how long it would take for this review of a specific set of rather general claims would be extended to "behavioural science was right!".

It didn't take long. Here's one statement from the lead author of the Nature paper:
